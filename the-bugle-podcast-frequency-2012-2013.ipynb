Let's analyze the podcasts produced at https://soundcloud.com/the-bugle/
and plot the number of full bugle podcasts released in a time-series.

This notebook is accessible [here](http://goo.gl/XxRh4j)

# Load the JSON produced by soundcloud

```python
import re
import json
import requests

%pylab --no-import-all inline

# pattern to distinguish Full-bugles versus Sub-bugles.
pat = re.compile('^Bugle (\d{3,4})(.*)')

# generated from soundcloud's API
url = 'https://dl.dropboxusercontent.com/u/42191402/projects2013/buglestat/all-tracks-clean.json'

req = requests.get(url)
js = req.json()

# Compensate for false positives in our regex above
to_filter_ids = [117136987]
full_episodes = [y for y in js if pat.match(y['title']) is not None and y['id'] not in to_filter_ids]
```

```python
import pprint
pprint.pprint(full_episodes[-1])
```

# Group full episodes by month

```python
from itertools import groupby
import numpy as np
import datetime
import matplotlib

groups = []
uniquekeys = []

keyfunc = lambda x: x['created_at'][0:7]
for k, g in groupby(full_episodes, keyfunc):
  groups.append(list(g))
  uniquekeys.append(k)

uniquekeys.reverse()
groups.reverse()

# convert for matplotlib's y-axis
yvals = np.asarray(map(lambda x: len(x), groups))
format =  "%Y/%m"

# x-axis 
dds = [datetime.datetime.strptime(x, format) for x in uniquekeys]
xvals = matplotlib.dates.date2num(dds)
print 'values on X-axis:\n', xvals
```

```python
print 'values on Y-axis:\n', yvals
```

# Plot the time-series

```python
f, ax = plt.subplots()
ax.plot_date(xvals, yvals, linestyle='-')
ax.grid(True)
plt.title('The Bugle Podcasts for 2012/2013')
plt.ylabel('Number of podcasts')
plt.ylim(0, 6)
plt.xlim(min(xvals)-60.0, max(xvals)+70.0)

# highlights
ax.annotate('Jan 2012 #179\n#savethebugle', (734503., 1.9), xytext=(0, -40), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

ax.annotate('#191 Chris\'s\nLondon marathon', (734594., 4.1), xytext=(-30, 40), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

ax.annotate('#200 London\nOlympics', (734685., 3.1), xytext=(-20, 40), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

ax.annotate('#193 Osama\ndeathiversary', (734624., 2.9), xytext=(-20, -40), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

ax.annotate('#224 Iron \nSheikh tweets', (734900., 4.1), xytext=(-50, 20), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

ax.annotate('Racist aussies\n& Thatcher #231', (734959., 1.9), xytext=(-10, -60), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

ax.annotate('#240 The \nAmerican returns', (735020., 4.1), xytext=(-40, 40), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

ax.annotate('#245 twerking', (735112., 3.1), xytext=(-20, 40), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

ax.annotate('Obamacare\n#250', (735173., 1.9), xytext=(-30, -40), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

ax.annotate('#209 Happy \n5 years!', (734747., 1.9), xytext=(0, -40), 
            textcoords='offset points', arrowprops=dict(arrowstyle='-|>'))

f.autofmt_xdate()
f.tight_layout()
plt.show()
```
